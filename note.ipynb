```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
y = np.array([0.055, 0.0275, 0.0183])
x1 = np.array([28.71, 10.94, 16.11])
x2 = np.array([143.75, 1077.77, 16.11])
x3 = np.array([261.11, 90, 22.43])

# Define the custom axis labels
custom_labels = {
    'x1': r'$\frac{X_A}{{(1 - X_A)^2}}$ (Temperature 17°C)',
    'x2': r'$\frac{X_A}{{(1 - X_A)^2}}$ (Temperature 25°C)',
    'x3': r'$\frac{X_A}{{(1 - X_A)^2}}$ (Temperature 32°C)',
    'y': r'$\tau$'
}

# Perform linear regression to find the best-fitting line
coefficients1 = np.polyfit(x1, y, 1)
coefficients2 = np.polyfit(x2, y, 1)
coefficients3 = np.polyfit(x3, y, 1)

# Calculate the slopes of the best-fitting lines
slope1 = coefficients1[0]
slope2 = coefficients2[0]
slope3 = coefficients3[0]

# Generate a range of x values for the fitted lines
x1_fit = np.linspace(min(x1), max(x1), 100)
x2_fit = np.linspace(min(x2), max(x2), 100)
x3_fit = np.linspace(min(x3), max(x3), 100)

# Calculate corresponding y values using the fitted lines
y1_fit = np.polyval(coefficients1, x1_fit)
y2_fit = np.polyval(coefficients2, x2_fit)
y3_fit = np.polyval(coefficients3, x3_fit)

plt.rcParams["figure.dpi"] = 360

# Plot the original data points and the best-fitting lines
plt.figure(figsize=(10, 6))

# Plot y vs x1
plt.scatter(x1, y, label='Original Data')
plt.plot(x1_fit, y1_fit,
         label=f'Best-Fitting Line (Slope: {slope1:.6f})', color='red')
plt.xlabel(custom_labels['x1'])
plt.ylabel(custom_labels['y'])
plt.title(r'$\tau$ vs $\frac{X_A}{{(1 - X_A)^2}}$ (Temperature 17°C)')
plt.legend()
plt.grid(True)
plt.show()

# Plot y vs x2
plt.figure(figsize=(10, 6))
plt.scatter(x2, y, label='Original Data')
plt.plot(x2_fit, y2_fit,
         label=f'Best-Fitting Line (Slope: {slope2:.6f})', color='red')
plt.xlabel(custom_labels['x2'])
plt.ylabel(custom_labels['y'])
plt.title(r'$\tau$ vs $\frac{X_A}{{(1 - X_A)^2}}$ (Temperature 25°C)')
plt.legend()
plt.grid(True)
plt.show()

# Plot y vs x3
plt.figure(figsize=(10, 6))
plt.scatter(x3, y, label='Original Data')
plt.plot(x3_fit, y3_fit,
         label=f'Best-Fitting Line (Slope: {slope3:.6f})', color='red')
plt.xlabel(custom_labels['x3'])
plt.ylabel(custom_labels['y'])
plt.title(r'$\tau$ vs $\frac{X_A}{{(1 - X_A)^2}}$ (Temperature 32°C)')
plt.legend()
plt.grid(True)
plt.show()
```
